我们用一个通俗的方式来理解它。

### 什么是文本情感分析？

想象一下，你有一个助手，他的工作不是帮你计算财报里的数字，而是**帮你“读懂”海量文字背后的人心**。

**1. 核心定义**
文本情感分析，又叫意见挖掘。简单说，就是**用计算机自动判断一段文字是“好评”还是“差评”，是乐观还是悲观，是愤怒还是喜悦**的过程。

如果把它放到**金融场景**里，它要解决的就是：**这篇文章、这条新闻、这份研报，对市场（或某只股票）到底是看涨、看跌，还是中性？**

### 举个栗子🌰

假设有三条关于某家上市公司的文本：

-   **文本A**：“这家公司的财报非常亮眼，营收远超预期，前景一片光明。”
    -   *情感：* **极度正面**（看涨信号）

-   **文本B**：“受宏观环境影响，该公司面临巨大的供应链挑战，预计下季度出货量将大幅下滑。”
    -   *情感：* **极度负面**（看跌信号）

-   **文本C**：“该公司今日召开了董事会，审议了年度预算方案。”
    -   *情感：* **中性**（陈述事实，无情绪）

对于人类来说，读完这几句话只需要几秒钟。但如果是**一万条**这样的文本，人类读不完。文本情感分析就是教计算机用极快的速度，把这“一万条”分成“好、坏、平”三类，甚至给它打一个**情感分数**（比如 -1 到 1 之间）。

### 它不只是“分正负”那么简单

在真正的金融研究中，情感分析通常分为三个层次：

1.  **篇章级 (Document Level)**：判断整篇文章是正面的还是负面的。
    -   *局限*：就像给整部电影打一个总分。如果一篇报道先夸了公司研发很强，后骂了公司现金流断裂，总分可能是“中性”，但这就丢失了关键信息。

2.  **句子级 (Sentence Level)**：判断每一个句子的情感。
    -   *应用*：更细致一些，但依然无法知道“研发很强”具体指的是哪个业务。

3.  **方面级 (Aspect Level)**：这是**金融分析中最实用**的层次。
    -   *逻辑*：它会拆解文本，针对特定主体判断情感。
    -   *例子*：“虽然**营收**增长迅猛（正面），但**现金流**极度恶化（负面）。”
    -   对于投资者来说，如果你只关注“现金流”这个方面，这个文本实际上是负面的。

### 为什么在金融数据分析中它很重要？

在你们的课程背景下，之所以要专门学习这个，是因为**传统的量化模型只看数字（股价、市盈率、交易量）**。但市场的大部分信息（政策解读、分析师预期、社交媒体情绪）最初都是以**文本**形式存在的。

文本情感分析提供了一种方法，把非结构化的“市场情绪”转化为结构化的“数据特征”。

-   **传统模型**：只能看到昨天股价跌了5%。
-   **加入情感分析后**：它能“看到”在股价跌5%的前一天，社交媒体上对该公司的负面情绪飙升了80%；或者是美联储的会议纪要里，关于“通胀”的担忧措辞比上个月增加了三倍。

### 总结

**文本情感分析，就是教计算机读懂“弦外之音”。**

在金融领域，它试图将“恐慌”、“贪婪”、“担忧”、“乐观”这些模糊的人类情绪，量化为 -0.5、+0.8 这样精确的数值，从而让你构建的量化模型，不仅能“看见”价格，还能“听见”市场的声音。

# 第 2 轮

>**我的提问**：先告诉我目前做情感分析的主流技术流派，他们有什么区别

```AI 的回答：```好的，我们进入第二个层次：**主流技术流派及其区别**。

理解流派，本质上是在理解情感分析这个任务是如何一步步从“人工规则”走向“机器直觉”的。目前主流的方法可以分为四代，它们在金融场景中各有优劣。



### 第一代：基于词典的方法

**原理**  
不涉及任何“学习”过程。它依赖一个预先构建好的**情感词典**（如正面词：增长、盈利、优秀；负面词：亏损、诉讼、暴跌）。算法扫描文本，统计正面词和负面词的数量，有时考虑否定词（如“不增长”）和程度副词（如“非常差”），最后计算一个情感得分。

**关键特点**
-   **无需训练数据**：拿来即用。
-   **可解释性极强**：你能明确指出是哪几个词导致了情绪判断。
-   **计算成本极低**。

**在金融领域的局限**
-   **领域迁移问题**：通用词典（如SentiWordNet）在金融文本上表现糟糕。比如“风险”在通用词典是负面的，但在风控报告中可能是中性；“做空”在通用词典是负面，但对对冲基金可能是正面操作。因此金融领域衍生出了专门的词典，如**Loughran-McDonald词典**（专门针对财报10-K文件构建）。
-   **无法理解复杂语法**：难以处理讽刺、条件句（“如果……那么……”）、隐含语义。



### 第二代：传统机器学习方法

**原理**  
将其视为一个分类问题（正面/负面/中性）或回归问题（情感得分）。需要**人工标注**一批数据作为训练集，从中提取**特征**（如词频TF-IDF、词性标签、句法结构），然后训练分类器（朴素贝叶斯、支持向量机SVM、逻辑回归等）。

**关键特点**
-   **需要标注数据**：质量取决于标注样本的数量和一致性。
-   **特征工程繁琐**：效果很大程度上取决于你能否设计出捕捉语义的特征。
-   **泛化能力有限**：在训练集分布相似的文本上表现良好，遇到全新表述方式容易失效。

**在金融领域的应用**
-   曾是主流。研究人员会标注数千条金融新闻，训练SVM模型。
-   比词典法更能捕捉组合特征，但仍然无法理解“这家公司虽然营收好，但我担心其现金流”这种转折与复杂语义。



### 第三代：深度学习（循环神经网络/CNN）

**原理**  
使用**词向量**（Word2Vec、GloVe）将词映射为稠密向量，通过**循环神经网络**（LSTM、GRU）或**卷积神经网络**（CNN）自动学习上下文特征。不再需要手动设计特征，模型从数据中端到端学习。

**关键特点**
-   **自动特征提取**：能捕捉词序和局部语义。
-   **需要大量数据**：相比传统机器学习，通常需要更多标注样本才能发挥优势。
-   **可解释性下降**：很难说清楚模型为什么判定一条新闻为负面。

**在金融领域的局限**
-   金融标注数据往往稀缺且昂贵（需要专家标注）。
-   LSTM虽然能捕捉序列，但对长距离依赖（比如跨段落推理）仍然吃力，且训练速度较慢。
-   随着Transformer架构出现，这类纯RNN/CNN模型在情感分析任务上已逐渐被预训练模型取代。



### 第四代：预训练语言模型（BERT及其金融变体）

**原理**  
这是目前**学术和工业界的事实标准**。模型在大规模通用语料（如维基百科、书籍）上进行**预训练**，学习了语言的通用语法和语义知识；然后在特定任务（金融情感分析）上用少量标注数据进行**微调**。

**关键特点**
-   **迁移学习能力**：用少量金融标注数据就能达到极佳效果。
-   **上下文感知**：解决了词典法“一词一义”的问题。比如“苹果”在“苹果发布新机”和“苹果价格下跌”中能根据上下文产生不同表征。
-   **计算成本高**：推理和训练需要GPU，模型较大（但蒸馏版本已可在普通机器上运行）。

**在金融领域的应用**
-   出现了大量金融领域预训练模型，如：
    -   **FinBERT**：在金融语料上继续预训练的BERT，对财报、新闻的情感分类效果显著优于通用BERT。
    -   **BloombergGPT**、**FinGPT**等大模型进一步提升了多任务能力。
-   当前金融情感分析的主流方案，往往是在FinBERT基础上微调少量标注数据。



### 第五代（新趋势）：大语言模型（LLM）提示工程

**原理**  
不进行微调，直接使用GPT-4、DeepSeek等大语言模型，通过设计**提示词**（“请判断以下新闻对某股票是利好、利空还是中性，并给出理由”）来抽取情感。

**关键特点**
-   **零样本/少样本能力**：甚至不需要标注数据，只需给出几个示例（in-context learning）。
-   **推理能力**：不仅能判断情感，还能给出逻辑链条（“因为提到了成本上升和需求下滑，所以是负面”），这对于金融决策非常宝贵。
-   **成本与延迟**：API调用成本高，延迟高，不适合实时处理千万级文本，但适合小规模分析或需要推理的场景。



### 核心区别总结（适合做幻灯片对比）

| 维度 | 词典法 | 传统机器学习 | 深度学习 (LSTM等) | 预训练模型 (FinBERT) | 大语言模型 (GPT) |

| **是否需要标注数据** | 否 | 是（千/万级） | 是（万级+） | 是（少量） | 否/极少 |

# 第 3 轮

>**我的提问**：如何构建一个简单的词典法模型

```AI 的回答：```好的，我们来一步步构建一个**基于词典的文本情感分析模型**。这是最透明、最容易上手的方法，非常适合作为金融数据分析课程的基线模型，而且你能清楚地看到每一个得分是怎么算出来的。



### 1. 整体思路

词典法模型的核心是：

1.  **准备一本情感词典**：每个词有一个情感极性（正面 +1，负面 -1），有时还有强度。
2.  **对文本进行预处理**：分词、去除标点、转换为小写等。
3.  **匹配词典**：统计文本中正面词和负面词的数量（或加权得分）。
4.  **可选增强**：考虑否定词（反转情感）、程度副词（加强或减弱）。
5.  **输出情感得分**：通常用 `(正面词数 - 负面词数) / 总词数` 或直接净得分。

在金融领域，最权威的词典是 **Loughran-McDonald 金融情感词典**（基于美国上市公司10-K年报构建），比通用词典（如 Bing Liu 词典）准确得多。



### 2. 准备词典

由于 Loughran-McDonald 词典是学术资源，你可以从作者的网站下载（`https://sraf.nd.edu/loughranmcdonald-master-dictionary/`）。但为了快速演示，我们先用一个简化的词典示例，包含常见金融词汇。实际作业中请替换为完整金融词典。

```python
])
negative_words = set([
    '下跌', '亏损', '诉讼', '风险', '下降', '疲软', '悲观', '低于预期', '利空', '恶化'

```

### 3. 文本预处理

我们写一个简单的预处理函数，将文本切分成词语列表。中文分词需要额外库，这里为了演示我们假设输入是中文句子，使用 `jieba` 分词（需要安装 `pip install jieba`）。如果处理英文则用 `nltk.word_tokenize`。

```python
import jieba
import re

def preprocess_text(text):
    # 去除标点符号和数字，保留中文和字母
    text = re.sub(r'[^\u4e00-\u9fa5a-zA-Z]', ' ', text)
    # 分词
    words = jieba.lcut(text)
    # 过滤掉单字词（可选）
    words = [w for w in words if len(w) > 1]
    return words
```



### 4. 基础情感得分计算

统计正面词和负面词的数量，计算净得分和归一化得分。

```python
def basic_sentiment_score(words, pos_set, neg_set):
    pos_count = sum(1 for w in words if w in pos_set)
    neg_count = sum(1 for w in words if w in neg_set)
    net_score = pos_count - neg_count
    # 归一化：除以总词数，避免文本长度影响
    if len(words) == 0:
        return 0.0
    normalized_score = net_score / len(words)
    return {
        'pos_count': pos_count,
        'neg_count': neg_count,
        'net_score': net_score,
        'normalized_score': normalized_score
    }
```



### 5. 加入否定词处理（进阶）

否定词（如“不”“没有”“并非”）会反转其后面词语的情感。简单做法是扫描文本，如果某个词的前一个词是否定词，则将其情感极性反转。

先定义一个否定词集合：

```python
negation_words = set(['不', '不是', '没有', '并非', '并未', '无'])
```

然后修改匹配逻辑，遍历词序列，记录前一个词是否是否定词。

```python
def advanced_sentiment_score(words, pos_set, neg_set, negation_set):
    pos_count = 0
    neg_count = 0
    i = 0
    while i < len(words):
        word = words[i]
        # 检查当前词是否被否定（前一个词是否在否定词集中）
        is_negated = (i > 0 and words[i-1] in negation_set)
        if word in pos_set:
            if is_negated:
                neg_count += 1   # 正面词被否定 → 负面效果
            else:
                pos_count += 1
        elif word in neg_set:
            if is_negated:
                pos_count += 1   # 负面词被否定 → 正面效果
            else:
                neg_count += 1
        i += 1
    net_score = pos_count - neg_count
    normalized_score = net_score / len(words) if words else 0.0
    return {
        'pos_count': pos_count,
        'neg_count': neg_count,
        'net_score': net_score,
        'normalized_score': normalized_score
    }
```

> **进一步**：还可以加入程度副词（“非常”“极度”“稍微”）来加权，但在简单模型中可先不实现。



### 6. 完整示例代码

我们将以上步骤整合，用几个金融新闻句子做测试。

```python
import jieba
import re

# 1. 简化的词典（实际应使用 Loughran-McDonald）
positive_words = set(['增长', '上升', '盈利', '收益', '优秀', '强劲', '乐观', '超预期', '利好', '改善'])

negation_words = set(['不', '不是', '没有', '并非', '并未', '无'])

def preprocess_text(text):
    text = re.sub(r'[^\u4e00-\u9fa5a-zA-Z]', ' ', text)
    words = jieba.lcut(text)
    words = [w for w in words if len(w) > 1]
    return words

def sentiment_score(words, pos_set, neg_set, negation_set=None):
    if negation_set is None:
        # 基础版本
        pos_count = sum(1 for w in words if w in pos_set)
        neg_count = sum(1 for w in words if w in neg_set)
    else:
        # 考虑否定词
        pos_count = 0
        neg_count = 0
        i = 0
        while i < len(words):
            word = words[i]
            is_negated = (i > 0 and words[i-1] in negation_set)
            if word in pos_set:
                if is_negated:
                    neg_count += 1
                else:
                    pos_count += 1
            elif word in neg_set:
                if is_negated:
                    pos_count += 1
                else:
                    neg_count += 1
            i += 1
    net_score = pos_count - neg_count
    normalized_score = net_score / len(words) if words else 0.0
    return {
        'pos_count': pos_count,
        'neg_count': neg_count,
        'net_score': net_score,
        'normalized_score': normalized_score
    }
```

### 一、什么是“预训练”？

**预训练**就像是让模型先接受一个“通识教育”。

-   **数据**：使用海量的、无标注的文本（比如整个维基百科、书籍、网页等）。
-   **目标**：让模型学习语言的通用知识——语法、句法、词义、上下文关系，甚至一些常识。比如，模型通过预训练会知道“苹果”可以是一种水果，也可以是一家公司，具体含义取决于它周围的词。
-   **方式**：通常通过“自监督学习”完成。比如 BERT 的做法是随机遮住文本中的一些词，让模型去猜被遮住的词是什么。这个任务不需要人工标注，所以可以用无限的数据。

**结果**：得到一个“基础模型”（如 BERT、FinBERT），它已经非常懂语言，但对具体任务（比如判断一段金融新闻是利好还是利空）还一无所知。



### 二、什么是“微调”？

**微调**则是在预训练的基础上，让模型接受“职业培训”。

-   **数据**：使用少量的、**有标注**的特定任务数据。比如，针对金融情感分析，你准备 1000 条新闻，每条标注好是“正面”“负面”还是“中性”。
-   **目标**：让模型从“懂语言”变成“懂金融情感分析”。微调时，模型会继续调整自己的所有参数，但调整幅度很小（因为基础已经很好），主要去学习如何把预训练获得的知识映射到你具体任务的标签上。
-   **方式**：在预训练模型的基础上，增加一个简单的输出层（比如一个分类头），然后用你的标注数据进行常规的监督学习（梯度下降）。

**结果**：得到一个专门用于金融情感分析的模型，它既保留了预训练阶段丰富的语言知识，又学会了你的任务所需要的特殊判断能力。



### 三、为什么只需要“少量”标注数据？

这是预训练模型相比传统机器学习最大的优势。

-   **传统机器学习**（如 SVM、朴素贝叶斯）没有预训练过程，它们完全依赖你提供的标注数据来学习。如果你的标注数据少，模型就学不到足够的特征，效果会很差。因此往往需要成千上万甚至更多的标注样本。
-   **预训练模型**已经通过海量无标注数据学会了语言的基本规律，微调只需要教会它“这个任务的目标是什么”。这个“适应”过程通常只需要几百到几千条标注数据就能达到非常好的效果。

-   **传统机器学习**：让你去一个全新的城市开一家餐厅。你对当地口味、食材供应链一无所知，需要花很多钱和时间做市场调研（标注数据）。

### 一、分类任务的标准评估指标

如果微调模型做的是情感分类（正面/负面/中性），这是典型的**多分类问题**。常用指标如下：

#### 1. 准确率 (Accuracy)
-   **定义**：预测正确的样本数 / 总样本数。
-   **局限**：类别不平衡时会失真。例如90%都是中性，模型全预测中性也有90%准确率，但完全没有识别能力。

#### 2. 精确率 (Precision)
-   **定义**：对每个类别，`预测为该类且正确的数量 / 预测为该类的总数量`。
-   **含义**：模型说“这是正面”时，有多大比例真的是正面。在金融中，如果你根据正面信号做多，高精确率能减少错误信号带来的亏损。

#### 3. 召回率 (Recall)
-   **定义**：对每个类别，`预测为该类且正确的数量 / 实际为该类的总数量`。
-   **含义**：所有真实的正面样本，模型找回了多少。召回率低意味着漏掉了很多机会。

#### 4. F1 分数
-   **定义**：精确率和召回率的调和平均数。  
    \( F1 = 2 \times \frac{Precision \times Recall}{Precision + Recall} \)
-   **适用**：当你希望同时兼顾精确率和召回率时使用。

#### 5. 宏平均与加权平均
-   **宏平均 (macro)**：先计算每个类别的指标，再取算术平均（所有类别权重相同）。
-   **加权平均 (weighted)**：按各类别样本数加权计算（更能反映整体表现）。

#### 6. 混淆矩阵 (Confusion Matrix)
-   最直观的工具：行是真实类别，列是预测类别。可以一眼看出模型容易混淆哪些类别（例如经常把“正面”误判为“中性”）。

**示例输出**（sklearn 风格）：
```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_true, y_pred, target_names=['negative', 'neutral', 'positive']))
print(confusion_matrix(y_true, y_pred))
```



### 二、回归任务的评估指标

如果你将情感映射为连续得分（例如 -1 到 1），则属于回归问题。常用指标：

-   **均方误差 (MSE)**：预测值与真实值差的平方的平均值，对异常值敏感。
-   **平均绝对误差 (MAE)**：绝对误差的平均值，更稳健。
-   **相关系数 (Pearson / Spearman)**：评估预测值与真实值的线性或单调相关性。在金融中，这比单纯的误差更有意义——你关心的是情感得分的相对排序是否准确。



### 三、稳健性评估

模型在测试集上表现好，不代表在真实世界中就可靠。还需要考察：

#### 1. 交叉验证
-   如果标注数据不多，建议使用 **K 折交叉验证**（如 5 折）。将数据分成 K 份，轮流做测试集，取平均指标，减少数据划分带来的偶然性。

#### 2. 过拟合判断
-   对比**训练集与验证集**的损失曲线。如果训练集损失持续下降而验证集损失上升，说明过拟合。
-   **早停法 (Early Stopping)** 可以在验证集不再提升时停止训练，是防止过拟合的有效手段。

#### 3. 不同领域/时间段的泛化能力

### 一、什么是 Loss（损失）？

**Loss 是一个数值，用来衡量“模型当前的预测有多离谱”。**

在情感分析任务中，假设你有一条新闻的真实标签是“正面”（我们用 one-hot 编码表示为 [0,0,1]），模型预测输出的概率分布是 [0.1, 0.2, 0.7]。损失函数（比如交叉熵）会计算这两个分布之间的差异——差异越大，loss 越大；模型越准，loss 越小。

**训练的目标就是让 loss 尽可能降低**，也就是让模型的预测不断逼近真实标签。



### 二、训练集 Loss vs. 验证集 Loss

在微调时，我们会把数据分成三部分：

- **训练集**：用来更新模型参数。
- **验证集**：不参与参数更新，只用来“旁观”模型在没见过的数据上表现如何。
- **测试集**：最后才用一次，评估最终模型的真实效果。

**典型现象**：
- **训练集 loss**：随着训练轮次（epoch）增加，会一直下降，因为模型在“背”训练数据。
- **验证集 loss**：一开始也会下降，但到了某个点之后，可能反而开始上升。

验证集 loss 开始上升的时刻，就是**过拟合**开始的信号——模型记住了训练集中的噪声，反而对没见过的数据预测能力下降。



### 三、什么是 Checkpoint？

**Checkpoint 是训练过程中，在某个时刻保存下来的模型参数快照。**

因为微调通常需要多轮迭代，你不可能等到训练全部结束才保存——万一后面过拟合了，或者训练中断，前面的好结果就丢了。所以常见的做法是：

- 每隔一定步数（例如每 100 步）或每个 epoch 结束时，保存一次模型。
- 每个保存点就是一个 checkpoint。



### 四、什么是“最佳 Checkpoint”？

“最佳 checkpoint”指的是：在**验证集上表现最好**的那个时刻的模型参数。

“表现最好”通常根据你关心的指标来衡量：
- 如果你关注分类准确度，可能看 **验证集 F1 分数**（或准确率）最高的那个 checkpoint。
- 如果你更关注概率校准，也可能选择 **验证集 loss 最低** 的那个 checkpoint。

**为什么不能直接取最后一个 epoch 的模型？**  
因为最后一个 epoch 的模型很可能已经过拟合——它在训练集上 loss 更低，但在验证集上 F1 可能已经下降。选择验证集上最佳的 checkpoint，能保证模型的**泛化能力**最强。



### 五、在代码中如何实现？

使用 Hugging Face `Trainer` 时，你只需要设置 `load_best_model_at_end=True` 和 `metric_for_best_model`，它就会自动帮你做这件事：

```python
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir='./results',
    evaluation_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    metric_for_best_model="f1",   # 或 "loss"
    greater_is_better=True,        # F1 越大越好，loss 则设为 False
)
```

如果没有用 Trainer，你也可以手动实现：
```python
best_f1 = 0
for epoch in range(num_epochs):
    train_one_epoch(model, train_loader)
    val_f1 = evaluate(model, val_loader)
    if val_f1 > best_f1:
        best_f1 = val_f1
        torch.save(model.state_dict(), "best_model.pt")
```

训练结束后，加载这个 `best_model.pt` 去测试集上做最终评估。

在金融场景中，你最终会用这个模型去分析未来的新闻、财报。如果模型过拟合了历史数据的噪声（比如某个特定时期“营收增长”总是伴随股价下跌），它在未来可能做出完全错误的判断，导致策略亏损。

选择验证集上表现最佳的 checkpoint，本质上是**在模型的拟合能力和泛化能力之间找到最佳平衡点**，确保你得到的不是“背题高手”，而是能真正理解金融文本情感的可靠工具。

# 第 8 轮

>**我的提问**：除了 F1，还有哪些指标适合作为选择 checkpoint 的依据？

```AI 的回答：```除了 F1 分数，选择最佳 checkpoint 时还有很多指标可供选择，具体取决于你的**任务类型（分类/回归）**、**数据特点**以及**金融应用的目标**。以下分几类展开，并说明在金融情感分析场景下的适用性。



### 一、分类任务的其他指标

#### 1. 验证集损失（Validation Loss）
-   **是什么**：模型在验证集上的交叉熵损失（或其他损失函数值）。
-   **优点**：直接反映模型预测概率分布的准确程度，对概率校准敏感，比离散指标（如准确率）更平滑。
-   **缺点**：损失值本身没有业务直观意义，难以横向比较。
-   **金融场景**：如果你后续需要将情感得分用于量化模型（作为输入特征），损失较低的模型往往能提供更可靠的概率估计，适合作为 checkpoint 选择依据。

#### 2. 宏平均 F1（Macro F1）
-   **是什么**：先分别计算每个类别的 F1，再取算术平均。
-   **适用**：类别不平衡时（金融文本中“中性”往往远多于“正面/负面”），宏平均 F1 会平等对待每个类别，避免模型忽略少数类。
-   **对比**：加权平均 F1 会偏向多数类。如果你关心识别“极端情绪”（如极负面新闻对市场的冲击），宏平均更合适。

#### 3. AUC-ROC（仅适用于二分类）
-   **是什么**：模型区分正负类的能力，与阈值无关。
-   **优点**：不受类别分布影响，衡量排序能力。
-   **扩展**：对于三分类（正/负/中），可使用 **宏平均 AUC**（一对多）或 **AUC-OVR**。
-   **金融场景**：如果你关心的是情感得分对未来的预测排序（例如做多得分最高的股票），AUC 比 F1 更贴近实际使用目标。

#### 4. 马修斯相关系数（MCC）
-   **是什么**：一种均衡的度量，取值 -1 到 1，可以理解为分类版的相关系数。
-   **优点**：在类别极不平衡时依然稳健，且能同时考虑所有混淆矩阵元素。
-   **金融场景**：如果数据中“负面”样本极少，MCC 能更真实反映模型对这类样本的识别能力。

#### 5. 平衡准确率（Balanced Accuracy）
-   **是什么**：各类别召回率的平均值。
-   **适用**：同样用于处理类别不平衡，直接告诉你模型在每个类别上的平均识别能力。



### 二、回归任务的指标（情感得分连续值）

如果你的微调模型输出的是连续情感分数（如 -1 到 1），则需要回归类指标：

#### 1. 平均绝对误差（MAE）
-   **优点**：直观、稳健，对异常值不敏感。
-   **金融场景**：如果你将情感得分作为特征，MAE 较小的模型在数值上更可靠。

#### 2. 均方误差（MSE） / 均方根误差（RMSE）
-   **优点**：对大误差惩罚更重，有助于避免极端错误预测。
-   **缺点**：对异常值敏感，解释性略差。

#### 3. 皮尔逊相关系数（Pearson）
-   **是什么**：衡量预测值与真实值的线性相关性。
-   **适用**：如果你关心情感得分的相对排序和趋势，相关系数比绝对误差更重要。

#### 4. 斯皮尔曼相关系数（Spearman）
-   **是什么**：基于排序的单调相关性，不要求线性关系。
-   **金融场景**：在选股策略中，你只需要情感得分的相对排名正确，斯皮尔曼相关系数比皮尔逊更适合作为选择依据。



### 三、金融领域特有的评估指标（进阶）

如果你希望 checkpoint 的选择更贴近最终应用目标，可以考虑直接使用金融效用指标——**但这需要格外小心**，因为直接用这些指标选择模型容易过拟合到验证集的特定时间段。

#### 1. 信息系数（IC，Information Coefficient）
-   **是什么**：情感得分与未来收益的斯皮尔曼相关系数。
-   **使用**：在验证集上计算 IC，选择 IC 最高的 checkpoint。这直接衡量了情感信号的预测能力。

#### 2. 策略回测指标（夏普比率、累计收益率）
-   **是什么**：根据情感得分构造简单交易策略（如得分 > 阈值买入），计算夏普比率或超额收益。
-   **风险**：这些指标方差很大，且可能过拟合验证集的特定行情。通常不推荐直接作为 checkpoint 选择依据，但可以作为辅助参考。



### 四、如何选择最适合的指标？

选择指标取决于你的最终目标：

| 最终应用 | 推荐选择指标 | 理由 |
|----------|--------------|------|
| 情感分类（报告展示） | **宏平均 F1** 或 **MCC** | 公平对待所有类别，避免多数类主导 |
| 情感得分作为量化模型输入特征 | **验证集损失** 或 **皮尔逊相关系数** | 损失反映概率校准，相关系数反映排序能力 |
| 情感得分用于构建选股信号 | **斯皮尔曼相关系数** | 排序正确性比绝对误差更重要 |
| 极少数类别（如“负面”）是关键 | **宏平均 F1**、**MCC** 或 **平衡准确率** | 确保模型不会忽视罕见但重要的情绪 |



### 五、实践建议

1. **至少监控两个指标**：一个反映整体性能（如验证集损失），一个反映业务关键性能（如宏平均 F1 或相关系数）。
2. **避免在验证集上过度优化**：如果你为了追求高 F1 而反复调整超参数，实际上会让验证集信息“泄漏”到模型选择中。建议在验证集上选 checkpoint，但最终用独立的测试集给出最终评估。
3. **结合时间序列划分**：金融数据有时间相关性。如果按时间划分验证集，选择验证集上最佳 checkpoint 可以一定程度上保证模型的时序泛化能力。

>**我的提问**：任务类型（分类/回归）有什么区别

```AI 的回答：```分类和回归是机器学习两大核心任务类型，在情感分析中对应两种不同的建模方式。它们的区别决定了你如何定义问题、选择模型、训练以及评估效果。下面从几个关键维度拆解。



### 一、本质区别

| 维度 | 分类 | 回归 |
|------|------|------|
| **输出** | 离散的类别标签 | 连续的数值 |
| **目标** | 判断样本属于哪个类别 | 预测一个数值，通常带有量级 |
| **例子** | 正面 / 中性 / 负面 | 情感得分 -0.82 ~ 0.73（连续值） |

**在情感分析中**：
- **分类任务**：给定一条金融新闻，判断它是“利好”“中性”还是“利空”。（三分类）
- **回归任务**：给这条新闻打一个情感分，比如从 -1（极度负面）到 +1（极度正面）的连续数值。



### 二、模型结构差异

#### 分类模型
- **输出层**：通常使用 Softmax 激活函数，输出每个类别的概率（例如 [0.1, 0.2, 0.7] 对应负/中/正）。
- **损失函数**：常用交叉熵损失（Cross-Entropy Loss）。
- **最后一层神经元数** = 类别数（如 3）。

#### 回归模型
- **输出层**：通常是一个神经元，不加激活函数（或使用线性激活），直接输出数值。
- **损失函数**：常用均方误差（MSE）或平均绝对误差（MAE）。
- **最后一层神经元数** = 1。

如果你使用预训练模型（如 FinBERT），只需要在模型顶部替换对应的输出头，就可以在分类和回归之间切换。



### 三、评估指标差异

#### 分类任务常用指标
- **准确率、精确率、召回率、F1 分数**（尤其是宏平均 F1 适合不平衡数据）
- **混淆矩阵**：直观看到错分情况
- **AUC-ROC**（二分类）或宏平均 AUC（多分类）

#### 回归任务常用指标
- **MAE、MSE、RMSE**：衡量预测值与真实值的数值误差
- **皮尔逊/斯皮尔曼相关系数**：衡量预测趋势与真实趋势的一致性（金融中很重要）



### 四、在金融情感分析中的应用场景

#### 什么时候用分类？
- **定性需求**：你只需要知道新闻是“利好”“利空”还是“中性”，比如用于事件驱动策略的信号（利好就买入，利空就卖出）。
- **标注数据为离散标签**：如果你获取的数据集本身标注了类别（如某些金融情感数据集），直接做分类更自然。
- **业务决策需要明确类别**：比如风控系统只关心“是否有负面情绪”，不需要程度。

#### 什么时候用回归？
- **需要情感强度**：你想捕捉“极度乐观”和“轻微乐观”的区别，例如在量化模型中作为因子（连续变量比离散变量携带更多信息）。
- **下游任务需要排序**：比如构建多空组合时，需要根据情感得分对所有股票排序，得分的高低比类别标签更重要。
- **标注数据是连续得分**：有些金融情感数据集提供的是情感分数（如分析师给出的情绪评分），此时回归更直接。



### 五、如何选择？

在课程作业中，你可以根据：
- **可用数据**：如果你的数据已经标注为三类，直接做分类；如果只有原始文本而没有标签，你可能需要先通过词典法生成连续得分作为伪标签，再用回归模型学习——但这样会引入噪声。
- **最终目的**：如果只是为了演示模型微调流程，分类更常见、评估更直观；如果想展示更深层的金融应用（如构建选股因子），回归往往更实用。
- **可解释性**：分类的结果（“利好”）更易懂，适合汇报；回归的结果需要解释得分的含义。

```AI 的回答：```对比分类与回归两种方式在同一数据上的效果差异，是一个很有价值的实验设计，能帮助你深入理解模型选择对结果的影响。下面提供一个系统性的对比框架，你可以直接用于作业中的实验部分。



### 一、实验设计思路

**前提条件**：你需要一个同时标注了**类别标签**和**连续情感得分**的数据集。如果原始数据只有连续得分，可以将其离散化为类别（例如：得分 > 0.3 为正面，< -0.3 为负面，其余为中性）来构建分类任务。

**核心思想**：
- 训练一个分类模型（如 FinBERT + 分类头），输出三个类别。
- 训练一个回归模型（如 FinBERT + 回归头），输出连续得分。
- 在相同的测试集上，用各自合适的指标评估，同时尝试将回归结果离散化后与分类结果对比，观察信息损失情况。



### 二、数据准备与预处理

#### 1. 数据集
假设你有一个金融新闻数据集，每条文本对应：
- **连续情感得分**（由人工标注或权威来源提供，范围 -1 到 1）
- **类别标签**（可以直接从连续得分映射得到）

映射规则示例：
- 正面：得分 ≥ 0.3
- 负面：得分 ≤ -0.3
- 中性：-0.3 < 得分 < 0.3

#### 2. 数据划分
- 按时间顺序划分（若有时序）或随机分层采样，确保训练、验证、测试集分布一致。
- 分类任务和回归任务使用完全相同的训练/验证/测试集划分，以保证对比公平。



### 三、模型构建与微调

#### 分类模型
- 使用预训练模型（如 FinBERT），顶部添加一个线性分类层，输出维度 = 3。
- 损失函数：交叉熵损失。
- 评估指标：验证集上选择最佳 checkpoint 的依据可以是**宏平均 F1**（适合类别不平衡）或验证集损失。

#### 回归模型
- 使用相同的预训练模型，顶部添加一个线性回归层，输出维度 = 1。
- 损失函数：均方误差（MSE）。
- 评估指标：验证集上选择最佳 checkpoint 的依据可以是**验证集 MSE** 或 **皮尔逊相关系数**（预测值与真实值的相关性）。



### 四、评估与对比

#### 1. 分类模型评估（在测试集上）
- 准确率、精确率、召回率、F1（宏平均和加权平均）
- 混淆矩阵（观察哪些类别容易混淆）
- AUC-ROC（若二分类可做，多分类可做一对多）

#### 2. 回归模型评估（在测试集上）
- MAE、MSE、RMSE
- 皮尔逊相关系数、斯皮尔曼相关系数

#### 3. 交叉对比方法
为了公平比较两种建模方式在“判断情感极性”上的能力，可以对回归模型的输出进行离散化（用同样的阈值映射为类别），然后计算其分类指标（准确率、F1 等），与直接分类模型对比。

1. 用回归模型预测测试集的连续得分。
2. 将预测得分按阈值转换为类别（例如 >0.3 → 正面，<-0.3 → 负面，其余中性）。

### 五、预期结果与分析

#### 1. 分类模型的预期表现
- **优势**：模型直接优化分类目标，在类别边界附近可能更准确；训练相对稳定，评估指标直观。
- **劣势**：可能忽略情感强度差异，将“极度正面”与“轻微正面”混为一谈；类别划分的阈值主观，若原始标签是连续值，离散化会损失信息。

#### 2. 回归模型的预期表现
- **优势**：保留情感强度信息，输出可用于排序和构建连续因子；在预测值与真实值的相关性上表现较好。
- **劣势**：若将回归结果离散化，其分类准确率可能低于直接分类模型，因为回归模型并未针对类别边界优化。

#### 3. 对比结果可能的发现
- **若回归模型离散化后的分类准确率接近甚至超过直接分类模型**，说明情感强度信息与类别划分高度一致，回归模型同样能学好边界。
- **若回归模型离散化后分类准确率显著低于直接分类模型**，说明分类模型在边界优化上更有优势，而回归模型可能为了降低 MSE 而在边界附近“妥协”。
- **回归模型的相关系数（尤其是斯皮尔曼）** 如果较高，说明其预测的情感排序与真实排序一致，这对于后续构建选股信号非常有利。

#### 4. 金融应用的启示
- 如果你只需要判断“利好/利空/中性”，直接训练分类模型可能更可靠。
- 如果你需要将情感作为因子（例如选股、风险模型），回归模型提供的连续得分通常更有价值，因为它能区分情绪的强弱。
- 在真实场景中，许多金融机构会先训练回归模型得到情感得分，再根据需要映射为类别——这样既保留了细粒度信息，又保持了灵活性。



### 六、实验局限性讨论

- **阈值的主观性**：离散化阈值的选取会影响对比结果，可尝试多个阈值（如 0.2、0.3、0.4）观察稳定性。
- **数据依赖性**：如果原始标签本身就是离散的（如人工标注的三分类），那么回归模型需要将类别转换为连续值（如 -1,0,1），这种人为编码可能引入噪声。
- **评估指标的可比性**：分类和回归的指标无法直接对比，需要借助“回归离散化”桥接。

下面我分几个层面来拆解这个问题。



### 一、什么是“复杂句子”中的情感识别难点？

复杂句子的情感识别之所以难，是因为它涉及超出“词袋”模型能力的语言现象：

1. **反讽/讽刺**：字面意思与真实意图相反。  
   > “真是好消息，公司股价一天跌了20%。”（真实情感：负面）

2. **转折与让步**：前半句正面，后半句负面，整体情感取决于重心。  
   > “虽然营收增长强劲，但现金流问题让人担忧。”（整体可能偏负面）

3. **条件与假设**：情感依附于未发生的条件。  
   > “如果美联储不加息，市场会大涨。”（情感不确定，取决于假设是否成立）

4. **隐含语义**：没有显性情感词，但通过事实陈述传达情绪。  
   > “该公司CEO在过去三年套现了80%的持股。”（隐含负面）

5. **金融专业表达**：  
   > “该股票被过度做空，可能面临轧空风险。”（“轧空”对散户是正面信号，对空头是负面）

|----------|--------------|----------------|--------------|

# 第 13 轮

>**我的提问**：有没有和文本情感分析相近的方法

```AI 的回答：```文本情感分析的核心是判断“作者的态度是正面还是负面”，但在实际研究和应用中，有一系列**相近但目标不同**的方法，它们要么与情感分析共享技术，要么在分析文本时互为补充。了解这些方法可以帮助你拓宽作业的分析维度，也能让你在金融文本挖掘中看到更完整的图景。



### 一、按任务目标分类

#### 1. 主题建模 (Topic Modeling)
- **做什么**：从大量文本中自动发现“在讨论什么话题”，不关心态度。
- **与情感分析的关系**：情感分析告诉你“情绪如何”，主题建模告诉你“为什么会有这种情绪”。两者常结合使用。
- **典型方法**：LDA（Latent Dirichlet Allocation）、BERTopic。
- **金融应用**：分析财报电话会议中管理层主要关注哪些话题（如“供应链”、“通胀”、“AI”），并观察各话题的情感倾向变化。

#### 2. 观点挖掘 (Opinion Mining) / 方面级情感分析 (Aspect-Based Sentiment Analysis)
- **做什么**：不仅判断整体情感，还定位到具体“方面”（aspect）并给出该方面的情感。
- **与情感分析的关系**：这是情感分析的细粒度版本。整体情感可能是“中性”，但具体到“现金流”是负面，“营收增长”是正面。
- **典型方法**：基于序列标注的模型（如BERT + CRF）、或使用大模型进行方面-情感对抽取。
- **金融应用**：分析分析师报告中对“估值”、“盈利”、“管理层”等不同方面的评价，构建细粒度因子。

#### 3. 立场检测 (Stance Detection)
- **做什么**：判断作者对某个特定目标（如某只股票、某项政策、某个公司）是支持、反对还是中立。
- **与情感分析的关系**：情感分析关注情绪，立场检测关注“对谁/对什么”的态度。情感分析可以没有明确目标，立场检测必须有目标。
- **典型方法**：目标-文本联合建模、基于提示学习的大模型。
- **金融应用**：检测市场对“美联储加息”的立场分布，或判断某条推文对“特斯拉”是看好还是看空。

#### 4. 意图识别 (Intent Detection)
- **做什么**：判断用户发表文本的意图（如询问、抱怨、推荐、决策）。
- **与情感分析的关系**：情感是情绪，意图是目的。一条愤怒的文本可能是“抱怨”意图，也可能是“威胁”意图。
- **典型方法**：分类模型（与情感分析共享技术），常与对话系统结合。
- **金融应用**：识别客户投诉邮件中的意图（要求赔偿、询问产品、解约），进行分类路由。

#### 5. 虚假信息检测 (Misinformation / Fake News Detection)
- **做什么**：判断文本内容是否真实可靠。
- **与情感分析的关系**：虚假新闻往往带有极端情感，两者特征有重叠，但目标不同。
- **典型方法**：结合文本特征、传播网络、用户画像的多模态模型。
- **金融应用**：识别社交媒体上针对上市公司的谣言，避免基于虚假信息的交易决策。



### 二、按技术路径分类（与情感分析共享技术栈）

#### 1. 文本分类 (Text Classification)
- **关系**：情感分析是文本分类的一个特例（按情感标签分类）。广义的文本分类还包括：主题分类、垃圾邮件检测、新闻类别分类等。
- **技术重叠**：几乎所有情感分析的技术（传统机器学习、BERT微调、大模型）都可直接用于其他文本分类任务。

#### 2. 文本回归 (Text Regression)
- **关系**：当情感分析输出连续得分时，它本质是文本回归。相似的回归任务包括：预测文本的“主观性分数”、“信息量分数”等。
- **技术重叠**：模型结构几乎一致，只是输出层和损失函数相同。

#### 3. 序列标注 (Sequence Labeling)
- **关系**：方面级情感分析中的“方面词抽取”属于序列标注（如命名实体识别 NER）。NER 是另一个与情感分析经常一起使用的技术。
- **技术重叠**：BERT + CRF 是经典结构，既可用于 NER，也可用于方面词抽取。

#### 4. 文本表示学习 (Text Representation Learning)
- **关系**：情感分析依赖于高质量的文本向量表示。同一套预训练模型（BERT、RoBERTa、FinBERT）也是其他所有文本任务的基石。
- **技术重叠**：无论做什么文本任务，预训练模型都是公共基础设施。



### 三、在金融分析中与情感分析互补的方法

| 方法 | 与情感分析的关系 | 金融典型应用 |
|------|------------------|--------------|
| **事件抽取** | 情感分析回答“情绪如何”，事件抽取回答“发生了什么事件” | 从新闻中抽取“并购”、“诉讼”、“高管离职”等结构化事件，再叠加情感分析判断市场反应 |

| **文本相似度 / 语义搜索** | 用于检索历史中与当前文本情感相似的案例 | 当某公司发布财报时，检索历史上情感分布相似的财报，观察其后股价走势 |
| **文本可读性分析** | 衡量文本复杂程度，与情感无关但常作为补充特征 | 财报可读性越低，往往隐藏负面信息；可读性与情感得分结合可提升预测能力 |

### 一、数据层面的局限与坑

| 问题 | 说明 | 金融场景举例 | 避坑建议 |
|------|------|--------------|----------|
| **情感的主观性与模糊性** | 同一段文本，不同读者可能产生不同情感判断。标注不一致会导致模型学习到噪声。 | 某新闻“公司宣布回购股票”——有人认为是利好，有人认为是管理层无更好投资机会的信号。 | 采用多人标注、明确标注指南；对于模糊样本，可标记为“中性”或直接剔除。 |
| **领域依赖性** | 通用情感词典（如正面词“good”）在金融领域可能失效。 | “风险”在通用词典是负面，但在风控报告里是中性的；“做空”在通用词典负面，但对冲基金视为正面操作。 | 使用金融专用词典（Loughran-McDonald）或基于领域语料微调模型。 |
| **数据不平衡** | 金融文本中“中性”样本往往占绝大多数（80%以上），极端情绪稀少。 | 财报新闻多数为中性陈述，只有少数重大事件才出现强烈正面或负面。 | 采用过采样、欠采样或使用加权损失；监控宏平均F1而非准确率。 |
| **时间漂移** | 文本的分布、词汇、情感含义会随时间变化。 | 2010年时的“区块链”可能是中性技术词，2021年则常与“泡沫”或“革命”等强烈情感绑定。 | 定期用新数据重新评估/微调模型；使用时间划分的验证集。 |



### 二、模型与算法层面的局限与坑

| 问题 | 说明 | 金融场景举例 | 避坑建议 |
|------|------|--------------|----------|
| **无法理解复杂语言现象** | 反讽、讽刺、双重否定、条件假设等仍对现有模型构成挑战。 | “真是好消息，股价一天跌了20%”——模型可能判为正面。 | 若数据中此类现象多，可考虑用大模型（GPT-4）或针对性数据增强；若数据多为正式文本，则影响不大。 |
| **缺乏常识与背景知识** | 模型仅基于文本，无法利用外部知识（如行业背景、公司历史）。 | “该公司CEO曾涉及会计丑闻” – 模型可能只将“丑闻”判为负面，但不知此CEO新上任带来的预期差。 | 结合知识图谱或结构化数据，采用多模态/多源融合。 |
| **忽略长文本结构** | 预训练模型有输入长度限制（通常512 token），财报或研报往往远超此长度。 | 10-K年报数万字，直接截断会丢失尾部信息或转折结构。 | 采用层次化模型（先分句/段处理再聚合），或使用Longformer、BigBird等长文本模型。 |
| **过拟合与泛化不足** | 在有限标注数据上微调，模型容易记住训练集噪声，而非真正学习情感规律。 | 训练集中“营收增长”总是正面，但实际中可能有“营收增长但利润下降”的复杂情况。 | 使用验证集早停；采用交叉验证；增加数据增强（回译、同义词替换等）。 |



### 三、评估与验证的坑

| 问题 | 说明 | 金融场景举例 | 避坑建议 |
|------|------|--------------|----------|
| **指标选择不当** | 用准确率评估不平衡数据会掩盖对少数类的识别能力。 | 模型把所有文本都判为中性，准确率90%却毫无用处。 | 使用宏平均F1、MCC、平衡准确率；对于回归任务使用相关系数而非仅MSE。 |
| **测试集泄露** | 在数据预处理（如归一化、特征选择）时混入了测试集信息。 | 对全文本做词频统计再划分训练/测试，导致测试集信息间接泄露。 | 所有预处理（包括分词词典构建）都应仅在训练集上进行，再应用到验证/测试集。 |
| **不考虑时间序列依赖** | 随机划分训练/测试集会破坏时间顺序，导致未来信息泄露到过去。 | 用2023年的数据预测2022年的情感，这在金融回测中是非法的。 | 严格按时间顺序划分：训练集时间早于验证集，验证集早于测试集。 |
| **忽略模型不确定性** | 只输出单一类别或得分，不提供置信度。 | 模型以0.51的概率判为正面，实则非常不确定，但使用者当作确定信号。 | 输出概率或置信区间；在策略中使用概率阈值或仅在高置信度时交易。 |



### 四、金融应用场景的特定坑

| 问题 | 说明 | 避坑建议 |
|------|------|----------|
| **情感与价格的因果关系陷阱** | 情感得分与股价可能相关，但因果方向难定：是情感驱动价格，还是价格影响情感？ | 使用格兰杰因果检验、事件研究法；注意领先滞后关系。 |
| **策略过拟合** | 在历史数据上优化情感阈值，得到完美回测，但未来失效。 | 使用滚动时间窗口验证；避免在测试集上反复调参；保持策略简单。 |
| **忽略新闻传播链** | 情感分析只处理文本内容，但市场反应还取决于新闻的发布渠道、时间、受众预期。 | 结合新闻来源权威性、首次发布与转发时间、覆盖度等元数据。 |
| **实时性要求** | 金融交易需要毫秒级响应，大模型推理延迟高。 | 对实时场景，采用轻量级模型（DistilBERT）或词典法；对大模型仅用于离线分析。 |
| **监管与合规风险** | 基于社交媒体情感的交易可能触及市场操纵红线。 | 避免基于虚假信息交易；使用公开、合法来源；咨询合规部门。 |

````python
import jieba
import re

# 1. 简化的词典（实际应使用 Loughran-McDonald）
positive_words = set(['增长', '上升', '盈利', '收益', '优秀', '强劲', '乐观', '超预期', '利好', '改善'])
negative_words = set(['下跌', '亏损', '诉讼', '风险', '下降', '疲软', '悲观', '低于预期', '利空', '恶化'])
negation_words = set(['不', '不是', '没有', '并非', '并未', '无'])

def preprocess_text(text):
    text = re.sub(r'[^\u4e00-\u9fa5a-zA-Z]', ' ', text)
    words = jieba.lcut(text)
    words = [w for w in words if len(w) > 1]
    return words

def sentiment_score(words, pos_set, neg_set, negation_set=None):
    if negation_set is None:
        # 基础版本
        pos_count = sum(1 for w in words if w in pos_set)
        neg_count = sum(1 for w in words if w in neg_set)
    else:
        # 考虑否定词
        pos_count = 0
        neg_count = 0
        i = 0
        while i < len(words):
            word = words[i]
            is_negated = (i > 0 and words[i-1] in negation_set)
            if word in pos_set:
                if is_negated:
                    neg_count += 1
                else:
                    pos_count += 1
            elif word in neg_set:
                if is_negated:
                    pos_count += 1
                else:
                    neg_count += 1
            i += 1
    net_score = pos_count - neg_count
    normalized_score = net_score / len(words) if words else 0.0
    return {
        'pos_count': pos_count,
        'neg_count': neg_count,
        'net_score': net_score,
        'normalized_score': normalized_score
    }

# 测试文本
texts = [
    "公司财报显示营收强劲增长，利润超预期，前景乐观。",
    "受行业低迷影响，该公司营收下降，利润出现亏损，风险加大。",
    "虽然营收增长，但净利润并未改善，反而恶化。"
]

for text in texts:
    words = preprocess_text(text)
    # 基础版本
    basic = sentiment_score(words, positive_words, negative_words)
    # 进阶版本（含否定词）
    advanced = sentiment_score(words, positive_words, negative_words, negation_words)
    print(f"文本：{text}")
    print(f"  分词：{words}")
    print(f"  基础得分：{basic['normalized_score']:.3f} (正:{basic['pos_count']}, 负:{basic['neg_count']})")
    print(f"  进阶得分：{advanced['normalized_score']:.3f} (正:{advanced['pos_count']}, 负:{advanced['neg_count']})")
    print()
````
